```python
import pandas as pd

df = pd.read_csv("D:\kaggle dataset\Sample - Superstore.csv",encoding="latin-1")
print(df.head())
print(df.info())
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df = df.sort_values(by="Order Date")
df
```

```python
#Sales Over Time (Trend)
import matplotlib.pyplot as plt

monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

```python
#Seasonal Pattern (Yearly, Quarterly, Monthly)
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()

sales_by_month = df.groupby("Month")['Sales'].sum().reindex(
    ["January","February","March","April","May","June","July","August","September","October","November","December"]
)

sales_by_month.plot(kind='bar', figsize=(12,6), title="Seasonal Sales Pattern (By Month)")
plt.show()
```

```python
#. Category & Sub-Category Sales Trends
import seaborn as sns

plt.figure(figsize=(12,6))
sns.lineplot(x="Order Date", y="Sales", hue="Category", data=df, ci=None)
plt.title("Sales Trend by Category")
plt.show()
```

```python
#Detect Anomalies (Using Rolling Average)
sales_daily = df.resample('D', on='Order Date')['Sales'].sum()
rolling_mean = sales_daily.rolling(window=30).mean()

plt.figure(figsize=(14,6))
plt.plot(sales_daily, label="Daily Sales", alpha=0.6)
plt.plot(rolling_mean, color="red", label="30-Day Rolling Avg")
plt.legend()
plt.title("Daily Sales with Rolling Average (Anomaly Detection)")
plt.show()
```

```python
#Profit vs Sales Trend
monthly_profit = df.resample('M', on='Order Date')['Profit'].sum()

plt.figure(figsize=(12,6))
plt.plot(monthly_sales.index, monthly_sales.values, label="Sales", marker='o')
plt.plot(monthly_profit.index, monthly_profit.values, label="Profit", marker='o')
plt.legend()
plt.title("Monthly Sales vs Profit Trend")
plt.show()
```
